```python
import numpy as np
import matplotlib.pyplot as plt

def vandermonde_interpolation(x, y, xi):
    """
    Perform polynomial interpolation using the Vandermonde matrix.
    x: array of input data points.
    y: array of function values corresponding to x.
    xi: points where interpolation is evaluated.
    """
    # Create the Vandermonde matrix
    V = np.vander(x, increasing=True)
    
    # Solve for coefficients of the polynomial
    coeff = np.linalg.solve(V, y)
    
    # Evaluate the polynomial at points xi
    yi = np.polyval(coeff[::-1], xi)
    return yi

# Functions to interpolate
def f1(x):
    return x**2

def f2(x):
    return np.sin(x)

def f3(x):
    return 1 / (1 + 2 * x**2)

# Main code
if __name__ == "__main__":
    # Define the interval and dataset sizes
    interval = [0, 2]
    dataset_sizes = [5, 10, 15, 20]  # Number of points in the dataset
    
    functions = [f1, f2, f3]
    function_names = ["f1(x) = x^2", "f2(x) = sin(x)", "f3(x) = 1/(1+2x^2)"]
    
    for func, name in zip(functions, function_names):
        print(f"\nAnalyzing interpolation for {name}:")
        
        for n in dataset_sizes:
            # Generate dataset
            x = np.linspace(interval[0], interval[1], n)
            y = func(x)
            
            # Points for evaluation
            xi = np.linspace(interval[0], interval[1], 1000)
            true_y = func(xi)
            
            # Perform interpolation
            yi = vandermonde_interpolation(x, y, xi)
            
            # Calculate error
            error = np.abs(true_y - yi)
            max_error = np.max(error)
            print(f"  Dataset size: {n}, Max Error: {max_error:.6f}")
            
            # Plot
            plt.figure(figsize=(8, 5))
            plt.plot(xi, true_y, label="True Function", color="blue")
            plt.plot(xi, yi, '--', label="Interpolated Polynomial", color="red")
            plt.scatter(x, y, color="green", label="Data Points")
            plt.title(f"Polynomial Interpolation for {name} (n={n})")
            plt.xlabel("x")
            plt.ylabel("y")
            plt.legend()
            plt.grid()
            plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def vandermonde_interpolation(x, y, xi):
    """
    Perform polynomial interpolation without using the Vandermonde matrix.
    x: array of input data points.
    y: array of function values corresponding to x.
    xi: points where interpolation is evaluated.
    """
    # Solve for coefficients of the polynomial using manual construction
    n = len(x)
    coeff = np.zeros(n)

    # Construct the polynomial coefficients by solving Ax = b
    for i in range(n):
        # Initialize the Lagrange basis polynomial
        basis = np.ones_like(x)
        for j in range(n):
            if i != j:
                basis *= (x - x[j]) / (x[i] - x[j])
        coeff[i] = np.dot(basis, y)

    # Evaluate the polynomial at points xi
    yi = np.zeros_like(xi)
    for i in range(n):
        term = coeff[i] * np.ones_like(xi)
        for j in range(n):
            if i != j:
                term *= (xi - x[j]) / (x[i] - x[j])
        yi += term

    return yi

# Functions to interpolate
def f1(x):
    return x**2

def f2(x):
    return np.sin(x)

def f3(x):
    return 1 / (1 + 2 * x**2)

# Main code
if __name__ == "__main__":
    # Define the interval and dataset sizes
    interval = [0, 2]
    dataset_sizes = [5, 10, 15, 20]  # Number of points in the dataset
    
    functions = [f1, f2, f3]
    function_names = ["f1(x) = x^2", "f2(x) = sin(x)", "f3(x) = 1/(1+2x^2)"]
    
    for i in range(len(functions)):
        func = functions[i]
        name = function_names[i]
        print(f"\nAnalyzing interpolation for {name}:")
        
        for n in dataset_sizes:
            # Generate dataset
            x = np.linspace(interval[0], interval[1], n)
            y = func(x)
            
            # Points for evaluation
            xi = np.linspace(interval[0], interval[1], 1000)
            true_y = func(xi)
            
            # Perform interpolation
            yi = vandermonde_interpolation(x, y, xi)
            
            # Calculate error
            error = np.abs(true_y - yi)
            max_error = np.max(error)
            print(f"  Dataset size: {n}, Max Error: {max_error:.6f}")
            
            # Plot
            plt.figure(figsize=(8, 5))
            plt.plot(xi, true_y, label="True Function", color="blue")
            plt.plot(xi, yi, '--', label="Interpolated Polynomial", color="red")
            plt.scatter(x, y, color="green", label="Data Points")
            plt.title(f"Polynomial Interpolation for {name} (n={n})")
            plt.xlabel("x")
            plt.ylabel("y")
            plt.legend()
            plt.grid()
            plt.show()
```
